## Date casting

````python
select('''
SELECT *
FROM users
WHERE
  created_at BETWEEN '2018-01-01' AND '2018-01-07'
ORDER BY created_at DESC;
''')
````

This query  won’t include signups from ‘2018-01-07’. ‘2018-01-07’ first will be converted to datetime type for comparison with created_at column: ‘2018-01-07 00:00:00’. This value points to the beginning of the day, so even the inclusive BETWEEN operator doesn’t help.


To improve the query we can convert (or type cast) the created_at column to date right there in the query:

````python
select('''
SELECT *
FROM users
WHERE
  created_at::date BETWEEN '2018-01-01' AND '2018-01-07'
ORDER BY created_at DESC;
''')
````

### Extracting date component


`DATE_TRUNC('datepart', timestamp)`

The ‘datepart’ parameter is a string from the list:

year
decade
century
quarter
month
week
day
hour
minute
second
millisecond


DATE_TRUNC() function always returns a datetime result, another timestamp which points to the beginning of the first day of a month or a year. We probably don’t want to see those days or zeros in hours/minutes/seconds positions. Especially when we need to extract such data for a report or presentation.


````python
sql = ''' 
SELECT
  DATE_TRUNC('year', created_at),
  COUNT(*)
FROM users
GROUP BY 1
ORDER BY 1 DESC
'''
select(sql)
````

`DATE_PART`

DATE_PART() function to extract a year from datetime field in just yyyy format


---

DATE_PART() function returns the number of a month. It’s not clear which year it is and if we have multiple years of data – we don’t really want to combine signups from multiple years into one group. ⚠ We need a function that extracts both year and month from datetime column.

---`

````python
sql = ''' 
SELECT
  DATE_PART('year', created_at),
  COUNT(*)
FROM users
GROUP BY 1
ORDER BY 1 DESC;
'''

select(sql)
````

`TO_CHAR(timestamp, ‘format’)`

The function we need is TO_CHAR(timestamp, 'format'). Note, that timestamp argument is now in the first position. The ‘format’ argument determines which parts of the timestamp and how will be extracted

```
symbol	meaning
yyyy	Year in 4-digit format, like 1988 or 2019
q	Quarter of a year, 1-4
mm	Number of month, 01-12
ww	Number of week within a year, 01-52
ddd	Number of day within a year, 01-365
dd	Number of day within a month, 01-31
hh	Hour of the day, 01-12
hh24	Hour of the day, 01-24
```

````python
sql = ''' 
SELECT
  TO_CHAR(created_at, 'yyyy-mm'),
  COUNT(*)
FROM users
GROUP BY 1
ORDER BY 1 DESC
'''
select(sql)
````

````python
sql = ''' 
    select 
        TO_CHAR(created_at, 'yyyy-mm') as month
        , count(id)
    from users 
    
    group by 1
    order by 2 desc
    limit 1;


'''
select(sql)
````

## DISTINCT

This is what happens when you run a SQL query:

- SQL engine first checks the table you specified in the FROM clause
- All the filters from the WHERE clause are applied
- It’s time to compute final values for the output in the SELECT clause (by this time we know that it could be function calls as well)
- If the query contains the DISTINCT statement all duplicated values will be filtered out.
- If the query contains the ORDER BY statement the records will be sorted.
- If the query contains LIMIT only a limited amount of records will be in the final result.

DISTINCT clause with multiple columns

We also can specify more than one column in the DISTINCT statement. In that case, only unique combinations of values will be included in the result


````python
select(
    ''' 
    select count(distinct country)
    from users 
'''
)
````

## UNIONS

UNION combines the results of 2 SQL queries together. There are a bunch of important features of UNION:

- result sets must have the same number of columns ⚠
- column types must match ⚠
- it filters out duplicated rows. The result of UNION will contain only unique rows 


UNION ALL

If we want to keep duplicated rows we can use UNION ALL:

````python
sql = ''' 

WITH emails AS (
  SELECT 
    'foo@bar.com' AS email,
    18 AS age
  UNION
  SELECT 
    'foo@example.org' AS email,
    18 AS age
  UNION
  SELECT 
    'bar@example.org' AS email,
    35 AS age
)

SELECT *
FROM emails

'''
select(sql)
````

## CASE statement

````python
sql = ''' 
WITH users_with_age_group AS (
  SELECT
    CASE
      WHEN age < 13 THEN 'kid'
      WHEN age < 18 THEN 'teenager'
      WHEN age < 25 THEN 'college'
      WHEN age < 35 THEN 'young adult'
      WHEN age < 56 THEN 'middle age'
      ELSE 'older adult'
    END AS age_group,
    *
  FROM users
  WHERE country = 'us'	
)

SELECT
  age_group,
  COUNT(*)
FROM users_with_age_group
GROUP BY 1

'''
select(sql)
````
